```python
import mysql.connector
import pandas as pd
from dotenv import load_dotenv
import os

# MySQL connection details are stored in connector.env file.
load_dotenv("connector.env")

# Establishing the connection to the MySQL database
conn = mysql.connector.connect(
    host = os.getenv("MYSQL_HOST"),
    user = os.getenv("MYSQL_USER"),
    password = os.getenv("MYSQL_PASSWORD"),
    database = os.getenv("MYSQL_DB")
)

table = "fragrances"

df = pd.read_sql(f"SELECT * FROM {table}", conn)

# Transforming DataFrame into csv file.
df.to_csv("fragrances.csv", index=False)

df.head()
```
